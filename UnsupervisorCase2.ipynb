1. Open and display the image “dog.jpeg”. Convert the image intoa numpy array,so that itcan beused in further processing.[Hint:Use the PIL module from python]
2. Find out the dimensions of the image and convert it intoa two-dimensional array.
3. Use K-meansclustering with k set to 3 and cluster the image.[Hint: Refer to k-means module of scikit learn]
4. Predict the cluster label of every pixel in the image and plot it back as an image.
5. Find out the three dominant colorsin the image.[Hint: The cluster centers should correspond to three dominant colors]


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load and display the image
image_path = "dogs.jpeg"
img = Image.open(image_path)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

# Convert to NumPy array
img_array = np.array(img)

# Reshape to 2D array
h, w, c = img_array.shape
pixels = img_array.reshape(-1, c)  # Reshape to (num_pixels, 3)

# Apply K-Means clustering
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixels)

# Get cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_.astype(int)  # RGB values of the centers

# Plot the clustered image
clustered_pixels = centers[labels]  # Map each pixel to its cluster center
clustered_image = clustered_pixels.reshape(h, w, c)  # Reshape back to original dimensions

plt.imshow(clustered_image)
plt.title("Clustered Image with K=3")
plt.axis("off")
plt.show()

# Extract dominant colors
dominant_colors = centers
print("Dominant Colors (RGB):")
for i, color in enumerate(dominant_colors, 1):
    print(f"Color {i}: {color}")
```
